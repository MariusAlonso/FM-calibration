```python
import pandas as pd

scores = pd.read_csv('calib_scores.csv', index_col=0)
scores.columns = ['dataset', 'metric', 'method', 'value']
scores['source'] = scores['dataset'].str.split('-').str.get(1)
scores['dataset'] = scores['dataset'].str.split('-').str.get(0)
scores = scores.dropna()
```

```python
scores.set_index(["dataset", "source", "metric", "method"]).unstack().dropna().xs(
    "ECE", level="metric"
).unstack("source").mean().unstack().T.style.background_gradient(axis=1)
```

```python
scores.set_index(["dataset", "source", "metric", "method"]).unstack().dropna().xs(
    "ECE", level="metric"
).rank(axis=1).unstack("source").mean().unstack().T.style.background_gradient(axis=1)
```

```python
print(
    scores.set_index(["dataset", "source", "metric", "method"])
    .unstack()
    .dropna()
    .xs("ECE", level="metric")
    .unstack("source")
    .mean()
    .unstack()
    .loc[:, ["CLIP", "TAB", "BERT", "GPT2", "MANTIS", "MOMENT"]]  # .rank(axis=1)
    .T.to_latex(float_format="%.3f")
)
```
